## 1

Пусть $x_0$ -- точка минимума. Тогда для всех $x$ выполнено:

$$ f(x) \ge f(x_0) = f(x_0) + \langle \mathbf{0}, x - x_0\rangle$$

То есть $\mathbf{0} \in \partial f(x_0)$.

Если же $\mathbf{0} \in \partial f(x_0)$, то для всех $x$ должно быть выполнено:

$$ f(x) \ge f(x_0) + \langle \mathbf{0}, x - x_0\rangle = f(x_0)$$

То есть $x_0$ -- точка минимума.

## 2

$f(x) = \text{ReLU}(x) = \max \{0, x\}$

По следствию из теоремы Моро-Рокафеллара о максимумах ($\partial 0 = \{0\},\; \partial x = \{1\}$):

$$
\partial f(x) =
\begin{cases}
0 & x < 0 \\
[0, 1] = \mathbf{conv}\{0, 1\} & x = 0 \\
1 & x > 0 \\
\end{cases}
$$

## 3

$f(x) = \|x\|_p$ при $p = 1,2, \infty$

1.$p = 1$

$$\|x\| = \sum_i |x_i|$$

По теореме Моро-Рокафеллара получим:

$$
[f_i(x)]_k =
\begin{cases}
|x_k| & k = i\\
0     & k \not = i
\end{cases}
$$

$$
\partial \|x\|=
\left\{ y \mid
\begin{cases}
y_i = \mathbf{sign}\ x_i & x_i \not = 0 \\
y_i \in [-1, 1] = \mathbf{conv}\{-1, 1\} & x_i = 0 \\
\end{cases}
\right\}
$$

2.$p = 2$

$$\|x\| = \sqrt{\sum_i x_i^2}$$

В $x_0 = 0$ посчитаем по определению, $f(x) = \|x\|$:

$$ \sqrt{\langle x, x\rangle} \sqrt{\langle y, y\rangle} \ge |\langle x, y \rangle| $$

$$ \sqrt{\sum_i x_i^2} \|y\| \ge f(x_0) + |\langle x - x_0, y \rangle| $$

$$ \sqrt{\sum_i x_i^2} \ge f(x_0) + |\langle x - x_0, \frac{y}{\|y\|} \rangle| \ge f(x_0) + \langle x - x_0, \frac{y}{\|y\|} \rangle$$

Причём равенство достигается при $x = ty, t \ge 0$. То есть множества $\partial \|x\|(0)$ и $\frac{y}{\|y\|}, \; y \in \mathbb{R}^n$ совпадают. Значит,

$$\partial \|x\|(0) = \{g \mid \|g\| \le 1\}$$

Записывая субдифференциал для произвольной точки (во всех, кроме $0$ есть градиент):

$$
\partial \|x\| =
\begin{cases}
\frac{x}{\sqrt{\sum_i x_i^2}} = \frac{x}{\|x\|} & x \not = 0 \\
G = \{g \mid \|g\| \le 1\} & x = 0
\end{cases}
$$

3.$p = \infty$

$$\|x\| = \max_i |x_i| $$

Рассмотрим эту норму как поточечный максимум выпуклых функций $f_i(x) = |x_i|$ на выпуклом множестве $\mathbb{R}^n$. Тогда 

$$
\partial \|x\| =
\mathbf{conv}\left\{\bigcup\limits_{\alpha \in \mathcal{A(x)}} \partial f_\alpha (x)\right\}, \; \mathcal{A(x)} = \{\alpha \mid x_\alpha = \max_i |x_i|\}
$$

$$ \partial f_\alpha(x)= \left\{ y \mid  
\begin{cases}
\begin{cases}y_i = \mathbf{sign} \ x_i & x_i \not = 0\\y_i \in [-1, 1] & x_i  = 0\\\end{cases}& i = \alpha \\
y_i = 0 & i \not = \alpha \\
\end{cases}
\right\}
$$

## 4

$f(x) = \|Ax - b\|_1^2$

Вводя обозначения: 

$$ h(x) = x^2, \; \mathbb{R} \to \mathbb{R}$$
$$ g(x) = \|x\|_1, \; \mathbb{R}^n \to \mathbb{R}$$
$$ u(x) = Ax - b, \; \mathbb{R}^m \to \mathbb{R}^n$$

Запишем:

$f(x) = h(g(u(x)))$

Для $h$ (монотонно неубывающая) и $u$ в каждой точке есть градиент, а $\partial g(x)$(монотонно неубывающая) посчитали ранее. По теореме о сложной функции получим:

$$\partial f(x) = h'(z) \rvert_{z=\|Ax - b\|_1} \partial g(u(x)) = 2\|Ax - b\|_1  \partial g(u(x))$$

$$
\partial u_i(x) = A_i = [i\text{-я строка матрицы }A] 
$$

$$
\partial g(u(x)) = \bigcup\limits_{p \in \partial g(z),\ z = u(x)} \left( \sum\limits_{i=1}^{n}p_i \partial u_i(x) \right) = \bigcup\limits_{p \in \partial g(z),\ z = u(x)} A^T p = A^T  \bigcup\limits_{p \in \partial g(z),\ z = u(x)}p
$$

Из ранее доказанного:

$$
\partial \|Ax - b\|_1=
\left\{ y \bigg{\rvert}
\begin{cases}
y_i = \mathbf{sign}\ [Ax - b]_i & [Ax - b]_i \not = 0 \\
y_i \in [-1, 1] = \mathbf{conv}\{-1, 1\} & [Ax - b]_i = 0 \\
\end{cases}
\right\}
$$

Собирая всё вместе:

$$ \partial f(x) = \left\{2\|Ax - b\|_1 A^T y \bigg{\rvert}
\begin{cases}
y_i = \mathbf{sign}\ [Ax - b]_i & [Ax - b]_i \not = 0 \\
y_i \in [-1, 1] = \mathbf{conv}\{-1, 1\} & [Ax - b]_i = 0 \\
\end{cases} 
\right\}$$

## 5 (не доделана)

$f(x) = e^{\|x\|}$

По теореме о субдифференциале сложной функции:

$$ \partial f(x) = e^{\|x\|} \partial \|x\|$$

Попробуем написать какие-нибудь оценки для $\partial \|x\|$.

Запишем определение:

$$\forall x\in S \quad \|x\| \ge \|x_0\| + \langle g, x-x_0\rangle$$

$$\forall x\in S \quad \langle g, x_0\rangle - \|x_0\| \ge \langle g, x\rangle - \|x\|$$

Пусть $x_0 = \alpha e_0$, $x = \beta e$, где $\|e\| = \|e_0\| = 1$, $\alpha \ge 0$, $\beta \ge 0$. Перепишем определение в виде:

$$\forall e \mid \|e\| = 1 \quad \forall \beta \ge 0 \quad \alpha(\langle g, e_0\rangle - 1) \ge \beta(\langle g, e\rangle - 1)$$

Пусть $x_0 = 0$, т.е. $\alpha = 0$:

$$\forall e \mid \|e\| = 1 \quad \forall \beta \ge 0 \quad 0 \ge \beta(\langle g, e\rangle - 1)$$

Так что для $x_0 = 0$ __получаем условие__:
$$\forall e \mid \|e\| = 1\quad \langle g, e\rangle \le 1 $$

$$\forall e \mid \|e\| = 1\quad \langle g, -e\rangle \ge -1 $$

$$\forall e \mid \|e\| = 1\quad \langle g, e\rangle \ge -1 $$

Осознаем, что это шар, сопряжённый к единичному шару $S = \{e \mid \|e\| = 1\}$ в данной норме в евклидовом пространстве со стандартным скалярным произведением:

$$S^* = \left\{g \mid \forall e \in S \quad \langle g, e\rangle \ge -1 \right\}$$

То есть $\partial \|x\| (0) = S^*$

Пусть теперь $x_0 \not = 0$, так что $\alpha > 0$. Тогда:

$$\forall e \mid \|e\| = 1 \quad \forall \beta \ge 0 \quad \langle g, e_0\rangle - 1 \ge \frac{\beta}{\alpha}(\langle g, e\rangle - 1)$$

Если $\langle g, e\rangle > 1$, то выбором $\beta$ мы можем сделать правую часть сколь угодно большой, что нарушит неравенство. Так что $\langle g, e\rangle \le 1$. Кроме того, если взять $e = e_0$, то получим:

$$ (1-\frac{\beta}{\alpha})(\langle g, e_0\rangle - 1) \ge 0$$

Выбором $\beta$ можно сделать $(1-\frac{\beta}{\alpha})$ как положительным, так и отрицательным, так что для справделивости неравенства требуется:

$$\langle g, e_0\rangle = 1$$

Итак, получили __следующие условия__, которые, как следует из рассуждений, эквивалентны определению при $x_0 \not = 0$:

$$\forall e \mid \|e\| = 1 \quad  \langle g, e_0\rangle = 1 \ge \langle g, e\rangle$$

Снова перефразируем через сопряжённое к единичному шару множество, но на этот раз надо учесть, что $ \langle g, e_0\rangle = 1 $. Заметим, что это условие задаёт некоторую прямую, перпендикулярную к $e_0$. То есть конструкция следующая:

$$ \partial \|x\| = S^* \cap L$$

$$ S = \{e \mid \|e\| = 1\}$$

$$ S^* = \{g \mid \forall e \in S \quad \langle g, e\rangle \ge -1 \}$$

$$ L   = \{g \mid \langle g, e_0\rangle = 1\}$$

Предположим, что эта прямая содержит точки из $\mathbf{int}S^*$. Тогда рассмотрим в ней точку $(g + \lambda e_0) \in S^*$, $\lambda > 0$. Должно быть выполнено:

$$\forall e \mid \|e\| = 1 \quad \langle g + \lambda e_0, e\rangle \ge -1$$

Взяв $e = -e_0$:

$$\langle g, -e_0\rangle + \lambda \langle e_0, -e_0\rangle \ge -1$$

$$ -1 - \lambda \langle e_0, e_0\rangle \ge -1$$

$$ \langle e_0, e_0\rangle \le 0$$

Но $e_0 \not = 0$, так что $\langle e_0, e_0\rangle > 0$, получили противоречие. Значит, прямая может содержать точки только из границы. То есть $L$ -- касательная к $S^*$ перпендикулярная к $e_0$.

__Важно:__ единичный шар может не быть самосопряжённым, ведь норма не согласована со стандартным скалярным произведением, взятым из определения субградиента. 

```python
import numpy as np
import matplotlib.pyplot as plt

N = 100
min_x, max_x = -2, 2
xs = np.linspace(min_x, max_x, N)

X, Y = np.meshgrid(xs, xs)

XY = np.stack((np.ravel(X), np.ravel(Y)), 1)

#Z = np.sum(np.fabs(XY), axis=1).reshape((N, N))
Z = np.sqrt(np.sum(XY**2, axis=1)).reshape((N, N))

pos = np.argwhere(Z <= 2)

S = np.stack((np.ravel(X[pos[:, 0], pos[:, 1]]), np.ravel(Y[pos[:, 0], pos[:, 1]])), 1)

S_c = np.all(np.dot(XY, S.T) >= -1, axis=1).reshape((N, N))

fig = plt.figure()

ax = fig.add_subplot(111)

ax.imshow(Z.reshape((N, N)) <= 2, cmap = 'Blues', interpolation='nearest', extent=(min_x, max_x, min_x, max_x), origin='lower', alpha=0.5)

ax.imshow(S_c, cmap = 'Reds', interpolation='bicubic', extent=(min_x, max_x, min_x, max_x), origin='lower', alpha=0.5)

ax.grid(True)
ax.axhline(0)
ax.axvline(0)
plt.tight_layout()
plt.show()
```
