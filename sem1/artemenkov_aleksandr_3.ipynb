## 1

Копипаста из разобранной задачи с оговорками на знак числа $\alpha$ и знаки неравенства:

* Из рисунка строим гипотезу: $\pi = y + \alpha c$. Коэффициент $\alpha$ подбирается так, чтобы $\pi \in S$: $c^T \pi = b$, т.е.: $$c^T (y + \alpha c) = b$$
$$c^Ty + \alpha c^T c = b$$
$$c^Ty = b - \alpha c^T c$$
Кроме того:
$$\alpha = \frac{b-c^T y}{c^Tc}$$

Так что:
* $y \in S \rightarrow \alpha \le 0$
* $y \not \in S \rightarrow \alpha > 0$



* Проверяем неравенство для выпуклого замкнутого множества. При $y \in S$ оно, очевидно, выполнено, и проекция совпадает с самой точкой. При $y \not \in S $ получаем $ \alpha > 0$ и далее: $(\pi - y)^T(x - \pi) \ge 0$
$$(y + \alpha c - y)^T(x - y - \alpha c) = $$
$$ \alpha c^T(x - y - \alpha c) = $$
$$ \alpha (c^Tx) - \alpha (c^T y) - \alpha^2 (c^Tc) \geq $$ (т.к. $\alpha > 0 $)
$$ \alpha b - \alpha (b - \alpha c^T c) - \alpha^2 c^Tc = $$
$$ \alpha b - \alpha b + \alpha^2 c^T c - \alpha^2 c^Tc = 0 \ge 0$$

## 2

Рассмотрим сначала более простую задачу, а именно проецирование на следующее множество $S$:

$X \in \mathbb{R}^{n \times m}$ и $S = \{x \in \mathbb{R}^n \mid x = X \alpha, \alpha \in \mathbb{R}^{m}\}$, $y \notin S$

Будем считать, что матрица $X$ невырождена. В противном случае всегда можно заменить матрицу $X$ на набор её базисных столбцов и породить то же самое линейное пространство $X \alpha$ (т.к. все остальные столбцы разложатся по базисным). Пусть матрица $F$ образована базисными векторами из ортогонального дополнения $X \alpha$, тогда выполнено: $X^TF = F^TX = 0$. Кроме того, произвольный $\mathbf{y} \in \mathbb{R}^n$ представим в виде:

$$\mathbf{y} = X\alpha_0 + F\beta_0$$

Умножим обе части равенства на $X^T$:

$$ X^T \mathbf{y} = X^TX\alpha_0 + X^TF\beta_0 = X^TX\alpha_0 $$

Т.к. матрица $X$ невырождена по построению, то $X^TX$ -- невырожденная квадратная матрица, так что она обратима. Откуда:

$$ (X^TX)^{-1}X^T \mathbf{y} = \alpha_0  $$

Окончательно, покажем, что проекция имеет вид:

$$ \pi_S(\mathbf{y}) = X\alpha_0 = X(X^TX)^{-1}X^T \mathbf{y}$$

Так как $X^TX$ -- симметричная, то $((X^TX)^T)^{-1} = ((X^TX)^{-1})^T$. Покажем, что

$$ (\pi_S(\mathbf{y}) - \mathbf{y})^T(\mathbf{x} - \pi_S(\mathbf{y})) \ge 0$$

$$ (X(X^TX)^{-1}X^T \mathbf{y} - \mathbf{y})^T(X\alpha - X(X^TX)^{-1}X^T \mathbf{y}) = $$
$$ (\mathbf{y}^TX((X^TX)^{-1})^TX^T - \mathbf{y}^T)(X\alpha - X(X^TX)^{-1}X^T \mathbf{y}) = $$
$$ (\mathbf{y}^TX(X^TX)^{-1}X^T - \mathbf{y}^T)(X\alpha - X(X^TX)^{-1}X^T \mathbf{y}) = $$
$$ \mathbf{y}^TX(X^TX)^{-1}X^TX\alpha - \mathbf{y}^TX\alpha - \mathbf{y}^TX(X^TX)^{-1}X^TX(X^TX)^{-1}X^T \mathbf{y} +  \mathbf{y}^TX(X^TX)^{-1}X^T \mathbf{y} = $$
$$ \mathbf{y}^TX\alpha - \mathbf{y}^TX\alpha - \mathbf{y}^TX(X^TX)^{-1}X^T \mathbf{y} +  \mathbf{y}^TX(X^TX)^{-1}X^T \mathbf{y} = 0 \ge 0$$

Теперь применим решение упрощённой задачи к исходной. Заметим, что взаимнооднозначное преобразование пространства $\phi(S) = S - x_0$ сохраняет расстояния, поэтому $\pi_{\phi(S)}(\phi(\mathbf{y})) = \phi(\pi_S(\mathbf{y}))$ (это следует из того, что $\rho (\phi(\mathbf{x}), \phi(\mathbf{y})) = \rho(\mathbf{x}, \mathbf{y})$ и определением проекции задаётся в обоих случаях одна и та же точка пространства, просто в разных базисах имеющая разные координаты). Иначе:

$$ \pi_S(\mathbf{y}) = \phi^{-1}(\pi_{\phi(S)}(\phi(\mathbf{y})))$$

Итак, получаем ответ:

$$ \pi_S(\mathbf{y})= X(X^TX)^{-1}X^T (\mathbf{y} - x_0) + x_0$$
