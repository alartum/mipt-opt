## 2

$S = \left\{ x \in \mathbb{R}^n_{++} \mid \prod\limits_{i=1}^n x_i \geq 1 \right\}$

Рассмотрим множество $S$ как надграфик функции $f(x): \; \mathbb{R}^{n-1}_{++} \rightarrow \mathbb{R}_{++}$
$$f(x) = \frac{1}{\prod\limits_{i=1}^{n-1}x_i}$$

$$\frac{\partial f(x)}{\partial x_i} = -\frac{1}{x_i^2 \prod\limits_{j \not = i} x_j}$$

$$\frac{\partial^2 f(x)}{\partial x_i \partial x_j} = 
\begin{cases}
\frac{2}{x_i^3 \prod\limits_{j \not = i} x_j} \text{ при } i = j\\
\frac{1}{x_i^2 x_j^2\prod\limits_{k \not \in \{i, j\}} x_j} \text{ при } i \not = j
\end{cases}$$

Все элементы гессиана на заданном множестве положительны, то есть для всех $x\in \mathbb{R}^{n-1}_{++} \; x^Tf''(x)x > 0$. Значит, функция выпукла, так что её надграфик -- выпуклое множество.

  Осталось заметить, что эта штука положительно определена, так что её слёд положителен.
 
* $g(X)=(\det X)^{\frac{1}{n}}$

Будем доказывать через ограничение на прямую. Во-первых, покажем, что функция $f(x) = (\prod\limits_{i=1}^{n}x_i)^{\frac{1}{n}}$ -- вогнута на $\mathbb{R}^n$. 

$$ \nabla f(x) = \frac{1}{n}f(x)\begin{pmatrix} \frac{1}{x_1} \\ \dotsc \\ \frac{1}{x_n} \end{pmatrix}$$

$$ \frac{\partial^2 f(x)}{\partial x_i \partial x_j} = \frac{1}{n^2}\frac{f(x)}{x_ix_j} -  \frac{1}{n}\frac{\partial x_i}{\partial x_j} \frac{f(x)}{x_i^2}$$

Запишем это равенство в матричном виде:

$$ f''(x) = \frac{f(x)}{n^2} r r^T -  \frac{f(x)}{n}\mathbf{diag}(x_1^2, \dotsc, x_n^2) $$

$$ y^Tf''(x)y = \frac{f(x)}{n^2} y^Tr r^Ty - \frac{f(x)}{n}y^T\mathbf{diag}(x_1^2, \dotsc, x_n^2)y$$

$$ y^Tf''(x)y = \frac{f(x)}{n^2} (r^Ty)^T (r^Ty) - \frac{f(x)}{n}y^T\mathbf{diag}(x_1^2, \dotsc, x_n^2)y$$

$$ y^Tf''(x)y = \frac{f(x)}{n^2} (r^Ty)^2 - \frac{f(x)}{n}y^T\mathbf{diag}(x_1^2, \dotsc, x_n^2)y$$

$$ y^Tf''(x)y = -(S^2 + R) \le 0$$

В $S$ собрали квадрат, в $R$ -- остальное. Так как работаем на положительных числах, то $R > 0$. Так что функция вогнута.

Отсюда следует, что и функция $u(x) = f(Ax + b) = (\prod\limits_{i=1}^{n}(\sum\limits_ja_{ij}x_i+b_i))^{\frac{1}{n}}$ -- вогнута. Теперь рассмотрим прямую: $x = t \begin{pmatrix}1 \\ \dotsc \\ 1\end{pmatrix}$. Получим:

 $$v(t) = (\prod\limits_{i=1}^{n}(\sum\limits_ja_{ij}t+b_i))^{\frac{1}{n}}$$

Но для данной пары матриц существует базис, в котором они обе имеют диагональный вид, причём $A \rightarrow E$. Тогда, раскладывая детерминант (диагональной) матрицы $A' + B'$, получим: 

$$g(X)= (\prod\limits_{i=1}^{n}(a_{ii}t+b_i))^{\frac{1}{n}}$$

Вогнута как суперпозиция вогнутыъ функций.

## 4

Покажем, что функция $f(p) = \sum\limits_{i=1}^n p_i \log p_i$ выпукла:

  $$ \nabla f(p) = \begin{pmatrix}1 + \log p_1\\ \dots \\ 1 + \log p_n \end{pmatrix} $$
    
  $$ \nabla^2 f(p) = \begin{pmatrix}\frac{1}{p_1} &  \dotsc & 0\\ \dots  & \dots & \dots  \\ 0  & \dotsc & \frac{1}{p_n}\end{pmatrix} $$
   
   $$\langle q, \nabla^2f(p)q\rangle  \ge 0 \; \forall q \in \mathbb{R}^n_{++}$$
   
   $$ q^T \begin{pmatrix}\frac{1}{p_1} &  \dotsc & 0\\ \dots  & \dots & \dots  \\ 0  & \dotsc &\frac{1}{p_n}\end{pmatrix} q = \sum\limits_{i=1}^{n}\frac{q_i^2}{p_i} \ge \frac{1}{p_{max}}\sum\limits_{i=1}^{n}q_i^2 \ge 0$$
   
   Значит, для неё выполнен дифференциальный критерий выпуклости первого порядка.
   
    $\forall p,q \in \mathbb{R}^n_{++}$:
$$
f(p) \ge f(q) + \nabla f^T(q)(p-q)
$$
$$
0 \le \sum\limits_{i=1}^{n}\left[p_i \log p_i - q_i \log q_i + (1 + \log q_i)(q_i-p_i)\right]
$$
$$
0 \le \sum\limits_{i=1}^{n}\left[p_i \log p_i - q_i \log q_i + q_i-p_i + q_i \log q_i  - p_i \log q_i\right]
$$
$$
0 \le \sum\limits_{i=1}^{n}\left[ p_i \log \frac{p_i}{q_i} + q_i-p_i\right]
$$

Более того, если задано ограниченное множество $S \subset \mathbb{R}^n_{++}$, то на нём функция сильно выпукла, так как $\frac{1}{p_{max}}$ принимает фиксированное значение и выполнен дифференциальный критерий выпуклости второго порядка. Значит, на любом ограниченном множестве:

   $$
f(p) \ge f(q) + \nabla f^T(q)(p-q) + \dfrac{\mu}{2}\|p-q\|^2
$$

   $$
f(p) - f(q) - \nabla f^T(q)(p-q) \ge \dfrac{\mu}{2}\|p-q\|^2
$$
  
$$ D(p,q) \ge \dfrac{\mu}{2}\|p-q\|^2
$$

Предположим, что существует $q \not = p: \; D(p, q) = 0$. Рассмотрим ограниченный $B_{r}(p): \; q \in B_{r}(p)$. Но тогда 

$$ D(p,q) \ge \dfrac{\mu}{2}\|p-q\|^2 > 0
$$

Получили противоречие, так что $D(p, q) = 0 \Rightarrow p = q$. В обратную сторону:

$$
D(p,p) = \sum\limits_{i=1}^n (p_i \log(p_i/p_i) - p_i + p_i) = 0
$$

Прежде всего удостоверимся, что множество $p$ -- выпукло. Действительно, оно является пересечением конечного числа выпуклых множеств: $(1 1 \dotsc 1)p = 1$ и $p_i \ge 0$.
  * $\mathbb{E}x$
  $$ \mathbb{E}x = \sum\limits_{i=1}^{n}a_ip_i = a^Tp = f(p)$$
  
  Убедимся, что для любых $p, q$ из множества и $\theta \in [0, 1]$:
  $$a^T(\theta p + (1- \theta)q) \le \theta a^Tp + (1- \theta)a^Tq$$
  
  Так что функция выпукла. То же самое неравенство можно записать и в другую сторону, так что функция и выпукла, и вогнута.
  * $\mathbb{P}\{x \ge \alpha\}$
  
  $$ \mathbb{P}\{x \ge \alpha\} = \sum\limits_{i=i_\alpha}^{n} p_i$$
  
  $$\sum\limits_{i=i_\alpha}^{n}(\theta p_i + (1- \theta)q_i) \le \theta \sum\limits_{i=i_\alpha}^{n}p_i + (1- \theta)\sum\limits_{i=i_\alpha}^{n}q_i$$
  
  Так что функция выпукла. То же самое неравенство можно записать и в другую сторону, так что функция и выпукла, и вогнута.
  
  * $\mathbb{P}\{\alpha \le x \le \beta\}$
  
   
  $$ \mathbb{P}\{x \ge \alpha\} = \sum\limits_{i=i_\alpha}^{i_\beta} p_i$$
  
  $$\sum\limits_{i=i_\alpha}^{i_\beta}(\theta p_i + (1- \theta)q_i) \le \theta \sum\limits_{i=i_\alpha}^{i_\beta}p_i + (1- \theta)\sum\limits_{i=i_\alpha}^{i_\beta}q_i$$
  
  Так что функция выпукла. То же самое неравенство можно записать и в другую сторону, так что функция и выпукла, и вогнута.
  
  * $\sum\limits_{i=1}^n p_i \log p_i$
 
 Введём $g_k(p_1, \dotsc, p_n) = f_k(p_k) =  p_k \log p_k$
  
  Так как функция $f(x) = x \log x$ выпукла, то и каждая из функций $g_k(p_1, \dotsc, p_n)$ выпукла. Но тогда исходная функция может быть получена как комбинация выпуклых функций с неотрицательными коэффициентами.
  
  $$\sum\limits_{i=1}^n p_i \log p_i = \sum\limits_{i=1}^n g_i(p_1, \dotsc, p_n)$$
  
   Так что функция выпукла. Более того она сильно выпукла:
   
    $$ \nabla f(p) = \begin{pmatrix}1 + \log p_1\\ \dots \\ 1 + \log p_n \end{pmatrix}$$
    
      
       
  $$ \nabla^2 f(p) = \begin{pmatrix}\frac{1}{p_1} &  \dotsc & 0\\ \dots  & \dots & \dots  \\ 0  & \dotsc & \frac{1}{p_n}\end{pmatrix} $$
   
   $$\langle q, \nabla^2f(p)q\rangle \geq \mu \|q\|^2$$
   
   $$ q^T \begin{pmatrix}\frac{1}{p_1} &  \dotsc & 0\\ \dots  & \dots & \dots  \\ 0  & \dotsc &\frac{1}{p_n}\end{pmatrix} q = \sum\limits_{i=1}^{n}\frac{q_i^2}{p_i} \ge \frac{1}{p_{max}}\sum\limits_{i=1}^{n}q_i^2 = \mu \|q\|^2$$
   
   Так как $p_{max} \le 1$, то выполнен дифференциальный критерий сильной выпуклости при $\mu = 1$. 
   
  * $\mathbb{V}x = \mathbb{E}(x - \mathbb{E}x)^2 = \mathbb{V}x = \mathbb{E}(x^2 - 2 x\mathbb{E}x + (\mathbb{E}x)^2) = \mathbb{E}x^2 - (\mathbb{E}x)^2$
  
  $$ \mathbb{V}x = \sum\limits_{i=1}^na_i^2p_i - \left(\sum\limits_{i=1}^{n}a_ip_i\right)^2 = f(p)$$
  
  $$ \nabla f(p) = \begin{pmatrix}a_1^2 - 2a_1\sum\limits_{i=1}^{n}a_ip_i \\ \dots \\ a_n^2 - 2a_n\sum\limits_{i=1}^{n}a_ip_i \end{pmatrix}$$
  
  $$ \nabla^2 f(p) = \begin{pmatrix}-2a_1^2 &  \dotsc & -2a_1a_n\\ \dots  & \dots & \dots  \\ - 2a_1a_n  & \dotsc & -2 a_n^2\end{pmatrix} = -2  \begin{pmatrix}a_1^2 &  \dotsc & a_1a_n\\ \dots  & \dots & \dots  \\ a_1a_n  & \dotsc & a_n^2\end{pmatrix} $$
  
  Проверяем:
  
  $$ p^T \nabla^2 f(p) p = -2\sum\limits_{k=1}^{n}p_k \sum\limits_{i=1}^{n}a_ka_ip_i = -2\sum\limits_{k=1}^{n}p_k a_k\sum\limits_{i=1}^{n}a_ip_i = -2 \left(\sum\limits_{k=1}^{n}p_k a_k \right)^2 \le 0$$
  
  Так что функция вогнута.
  * $\mathbf{quartile}(x) = {\operatorname{inf}}\left\{ \beta \mid \mathbb{P}\{x \le \beta\} \ge 0.25 \right\}$
  
  Покажем, что функция не является ни выпуклой, ни вогнутой. Рассмотрим пример для $\mathbb{R}^2$:
  
  $$p = \begin{pmatrix}p_1 \\p_2 \end{pmatrix}, \quad p_1 + p_2 = 1, \quad p_1 \ge 0, \quad p_2 \ge 0$$
  
  Возьмём точки 
  
  $$ p = \begin{pmatrix}\frac{1}{2} \\ \frac{1}{2}\end{pmatrix}, \quad q = \begin{pmatrix}\frac{1}{8} \\ \frac{7}{8}\end{pmatrix}$$
  
  Тогда по определению: 
  
  $$ f(p) = a_1, \quad f(q) = a_2, \quad a_1 < a_2$$
  
  Теперь возьмём $\theta_1 = \frac{1}{2}$:
  
 $$ \theta_1 p + (1 - \theta_1)q = \frac{1}{2} \cdot  \begin{pmatrix}\frac{1}{2} \\ \frac{1}{2}\end{pmatrix} + \frac{1}{2}\cdot \begin{pmatrix}\frac{1}{8} \\ \frac{7}{8}\end{pmatrix} = \begin{pmatrix}\frac{5}{16} \\ \frac{11}{16}\end{pmatrix} = r_1$$
 
 $$ f(\theta_1 p + (1 - \theta_1)q) = f(r_1) = a_1 = \frac{1}{2} a_1 + \frac{1}{2}a_1 < \frac{1}{2} a_1 + \frac{1}{2}a_2 = \theta_1 f(p) + (1-\theta_1)f(q)$$
 
 Если же взять $\theta_2 = \frac{1}{4}$:
 
  $$ \theta_2 p + (1 - \theta_2)q = \frac{1}{4} \cdot  \begin{pmatrix}\frac{1}{2} \\ \frac{1}{2}\end{pmatrix} + \frac{3}{4}\cdot \begin{pmatrix}\frac{1}{8} \\ \frac{7}{8}\end{pmatrix} = \begin{pmatrix}\frac{7}{32} \\ \frac{25}{32}\end{pmatrix} = r_2$$
 
 $$ f(\theta_2 p + (1 - \theta_2)q) = f(r_2) = a_2 = \frac{1}{4} a_2 + \frac{3}{4}a_2 > \frac{1}{4} a_1 + \frac{3}{4}a_2 = \theta_2 f(p) + (1-\theta_2)f(q)$$
 
 Так что на этом отрезке существуют две точки, в которых неравенство выполнено в разные стороны. Так что выпуклости нет.
